```python
import numpy as np

# esta seria la primera fila de V
# copias la matriz V aqui y ya pruebas con cada vector b, el resultado tiene que ser la matriz x
vector_fila_v = np.array([
    [0.11, -0.8, 0.3, -0.17, -0.48],
    [-0.58, -0.31, 0.01, -0.5, 0.56],
    [0.49, 0.08, -0.49, -0.72, -0.07],
    [0.6, -0.36, 0.04, 0.27, 0.66],
    [-0.23, -0.36, -0.82, 0.37, -0.09]])
    
vector_columna_b = np.array([0, -3.2, 0, 0, 0])
vector_columna_u = np.array([12.9, 58.2, 1.7, 1436.8, 4.1])

# Multiplicaci√≥n de matrices
resultado = np.matmul(vector_fila_v, vector_columna_b) + vector_columna_u
print("matriz x")
print(resultado)
```
