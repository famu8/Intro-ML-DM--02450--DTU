```python
import numpy as np

x = np.array([-0.1, 0.2, 0.1, -0.3, 1, 0.5])
pc1 = np.array([0.38, 0.41, 0.5, 0.29, 0.45, 0.39])
pc2 = np.array([-0.51, 0.41, 0.34, 0.48, -0.42, -0.23])



cord1 = np.dot(x.T,pc1)
print(cord1)

cord2 = np.dot(x.T,pc2)
print(cord2)
```
