```python
"""
En este ej te van a dar varios boundaries y tienes que ir comprobando uno a uno.
Aqui abajo tienes las distancias
"""
```

```python
import numpy as np

# DISTANCIA 1
# Definir dos vectores columna
v1 = np.array([[-1.24], [-0.26], [-1.04]])
v2 = np.array([[-0.6], [-0.86], [-0.5]])

# Calcular la distancia de norma 1
distance = np.sum(np.abs(v1 - v2))

# Imprimir el resultado
print("Distancia de norma 1:", distance)
```

```python
# DISTANCIA 2

# Definir dos vectores columna
v1 = np.array([[2.7], [2.8]])
v2 = np.array([[6], [0]])

# Calcular la distancia de norma 2
distance = np.linalg.norm(v1 - v2)

# Imprimir el resultado
print("Distancia de norma 2:", distance)
```

```python
# DISTANCIA 3
# Definir dos vectores columna
v1 = np.array([[2], [3]])
v2 = np.array([[6], [3]])

# Calcular la norma 3
distance = np.power(np.sum(np.abs(v1 - v2)**3), 1/3)

# Imprimir el resultado
print("Norma 3:", distance)
```

```python
import numpy as np

# Definir dos vectores columna
v1 = np.array([[-1.24], [-0.26], [-1.04]])
v2 = np.array([[-0.6], [-0.86], [-0.5]])

# Calcular la norma 4
distance_4 = np.power(np.sum(np.abs(v1 - v2)**4), 1/4)

# Calcular la norma 5
distance_5 = np.power(np.sum(np.abs(v1 - v2)**5), 1/5)

# Imprimir los resultados
print("Norma 4:", distance_4)
print("Norma 5:", distance_5)
```

```python
# DISTANCIA INF

# Definir dos vectores columna
v1 = np.array([[-1.24], [-0.26], [-1.04]])
v2 = np.array([[-0.6], [-0.86], [-0.5]])

# Calcular la norma infinito
distance = np.max(np.abs(v1 - v2))

# Imprimir el resultado
print("Norma infinito:", distance)
```
