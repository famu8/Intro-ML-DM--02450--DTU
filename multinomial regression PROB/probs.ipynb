```python
"""
Te van a dar los pesos 
Te van a dar dos valores de b y tienes que crear un vector columna: 
En primera posicion un 1 y luego b1 y b2...
Luego para calcular y tienes que hacer
y=b.T * w1 (que te lo dan)

Luego simplemente aplicas las formula:
p(...)=math.e(y1) / 1 + math.e(y1) + math.e(y2)
Esto te dar el primer valor de la primera posicion de
uno de los vectores que te dan como resutado 
"""
```

```python
import numpy as np
import math
b=np.array([1, -5.52, -4.69])
w1=np.array([0.04, 1.32, -1.48])
w2=np.array([-0.03, 0.7, -0.85])
# ojo b antes siempre
y1=np.dot(b.T,w1)
y2=np.dot(b.T,w2)

# calculas la prob
prob = math.exp(y1) / (1 + math.exp(y1) + math.exp(y2))
prob
```
