```python
def simple_matching_coefficient(frase1, frase2):
    # Tokeniza las frases en palabras
    palabras_frase1 = set(frase1.split())
    palabras_frase2 = set(frase2.split())
    
    # Encuentra la intersección de las palabras (coincidencias)
    interseccion = len(palabras_frase1.intersection(palabras_frase2))
    print("numero palabras que se repiten: ",interseccion)
    
    # Calcula el número total de palabras en ambas frases (K)
    palabras_unicas_frase1 = set(frase1.split())
    palabras_unicas_frase2 = set(frase2.split())
    K = len(palabras_unicas_frase1.union(palabras_unicas_frase2))
    print("Total de frase:", K)
    
    # Calcula el Simple Matching Coefficient (SMC)
    smc = interseccion / K
    
    return smc

# Ejemplo de uso
frase1 = "the cruise will take your breath away"
frase2 = "wash your hands before the buffet"
smc = simple_matching_coefficient(frase1, frase2)
print(f"Simple Matching Coefficient (SMC): {smc:.7f}")
```
