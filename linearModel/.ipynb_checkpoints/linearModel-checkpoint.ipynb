```python
X = np.array([[1,1],
    [1,3],
    [1,4]])
y = np.array([2,5,6])
```

```python
term1 =  np.dot(X.T, X)
```

```python
term1
```

```python
term1_inverse = np.linalg.inv(term1)
```

```python
term1_inverse
```

```python
term2 = np.dot(X.T,y)
```

```python
w = np.dot(term1_inverse, term2)
```

```python
# como te dicen que quieren la muestra x=5 pues haces lo siguiente
pred = np.array([1, 5])
```

```python
y_final = np.dot(pred,w)
```
